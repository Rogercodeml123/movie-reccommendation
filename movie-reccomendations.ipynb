```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movis['genres']=movis['genres'].apply(convert)
```

```python
movis.head()
```

```python
movis['keywords']=movis['keywords'].apply(convert)
```

```python
#movis['cast'][0]  
# we  need fst three actual names of a actor 
```

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movis['cast']=movis['cast'].apply(convert3)
```

```python
movis.head()
```

```python
#movis['crew'][0]
# we need job value as director in every rows dictionary
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movis['crew']=movis['crew'].apply(fetch_director)
```

```python
#vectorization of movie_id,title,tags and we have 5000 movies 
# text vectorization = we convert every text in a vector then every movie became  a vector
# like (2,3)  is a movie that we like and we have 5000 vectors in our space 
# then we find nearest 5 movies vectors that is similar to our liked movie 
#we find cosine distance for find nearest similar movie by theta
#techniques = bag of words , tf-idf, wordtovec
# bag of words =  combine all words in tags i.e. tag1+ tag2....lakh words
# we need 5000 most common words which have more frequency
#    word1/w2/w3/w4.......w5000
# m1:     5,3,0,1,2,........1
# m2:     2,0,1,............3
# .       ..............
# m5000   5,2,5,...........5
#now every movie is a vector for 5000 dimen. space
#(5000,5000)     if 2d space then two words for 5000 movies
# have to find best perf. in less words so that dimenssionality not inc so not 10k words
# cant consider stop words : not contribution in sentence meaning i.e. are ,in etc

```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def  stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```
